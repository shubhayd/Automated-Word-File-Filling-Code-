```python

# ---- Auto-fill {{placeholders}} in a Word document ----
from docx import Document

def replace_placeholders_in_paragraph(paragraph, mapping):
    """
    Replace placeholders in a paragraph. If a placeholder spans multiple runs,
    this collapses runs and may lose inline formatting around the replaced text.
    """
    text = paragraph.text
    replaced = False
    for placeholder, value in mapping.items():
        if placeholder in text:
            text = text.replace(placeholder, str(value))
            replaced = True

    if replaced:
        # Clear existing runs and replace with a single run containing the new text
        for run in paragraph.runs[::-1]:
            paragraph._element.remove(run._element)
        paragraph.add_run(text)

def replace_placeholders_in_docx(template_path, output_path, mapping):
    """
    mapping: dict where keys are placeholders like '{{ApplicantName}}'
             and values are the replacement strings.
    """
    doc = Document(template_path)

    # Replace in body paragraphs
    for paragraph in doc.paragraphs:
        replace_placeholders_in_paragraph(paragraph, mapping)

    # Replace inside tables
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for paragraph in cell.paragraphs:
                    replace_placeholders_in_paragraph(paragraph, mapping)

    # Replace in headers/footers
    for section in doc.sections:
        header = section.header
        footer = section.footer
        for paragraph in header.paragraphs:
            replace_placeholders_in_paragraph(paragraph, mapping)
        for paragraph in footer.paragraphs:
            replace_placeholders_in_paragraph(paragraph, mapping)

    doc.save(output_path)

if __name__ == "__main__":

    # Ask user for input
    Date_of_document = input("Enter Document Date: ")
    Duration_of_Workshop = input("Enter Duration of Workshop: ")
    Title_of_Workshop = input("Enter Title of Workshop: ")
    Location_of_Workshop = input("Enter Location of Workshop: ")
    Date_of_Workshop = input("Enter Date of Workshop: ")
    Optimistic_number = input("Optimistic number: ")
    Pessimistic_number = input("Pessimistic number: ")
    Eng_and_prof_fee = input("Engineers & Professionals Fees: ")
    Total_Receipts_Optimistic_Rs_ = input("Total Receipts Optimistic Rs_: ")
    Total_Receipts_Pessimistic_Rs_ = input("Total Receipts Pessimistic Rs_: ")
    Teaching_faculty_fee = input("Teaching Faculty Fee Rs_: ")
    Eng_college_student_fee = input("Eng college student Fee Rs_: ")
    Arai_acy_student_fee = input("ARAI Acy student fee Rs_: ")
    Total_Receipts_Optimistic = input("Total Receipts Optimistic Rs_: ")
    Total_Receipts_Pessimistic = input("Total Receipts Pessimistic Rs_: ")
    Optimistic_Manpower_cost_Rs_ = input("Optimistic Manpower cost Rs_: ")
    Pessimistic_Manpower_cost_Rs_ = input("Pessimistic Manpower cost Rs_: ")
    Notional_percentage = input("Notional %: ")
    Optimistic_Notional_Rs_ = input("Optimistic Notional Rs_: ")
    Pessimistic_Notional_Rs_ = input("Pessimistic Notional Rs_: ")
    Optimistic_hall_charges_Rs_ = input("Optimistic hall charges Rs_: ")
    Pessimistic_hall_charges_Rs_ = input("Pessimistic hall charges Rs_: ")
    Honorarium = input("Honorarium Rs_: ")
    Transport_Speaker_Rs_Approx = input("Transport Speaker Rs Approx: ")
    Optimistic_Honorarium_Rs_ = input("Optimistic Honorarium Rs_: ")
    Pessimistic_Honorarium_Rs_ = input("Pessimistic Honorarium Rs_: ")
    Optimistic_Delegate_kits_Rs_ = input("Optimistic Delegate kits Rs_: ")
    Pessimistic_Delegate_kits_Rs_ = input("Pessimistic Delegate kits Rs_: ")
    Optimistic_Catering_Expenses = input("Optimistic Catering Expenses Rs_: ")
    Pessimistic_Catering_Expenses = input("Pessimistic Catering Expenses Rs_: ")
    Optimistic_Miscellaneous_expenses = input("Optimistic Miscellaneous expenses Rs_: ")
    Pessimistic_Miscellaneous_expenses = input("Pessimistic Miscellaneous expenses Rs_: ")
    Optimistic_total_expense = input("Optimistic total expense Rs_: ")
    Pessimistic_total_expense = input("Pessimistic total expense Rs_: ")
    Optimistic_Surplus_of_expense = input("Optimistic Surplus of expense Rs_: ")
    Pessimistic_Surplus_of_expense = input("Pessimistic Surplus of expense Rs_: ")

    # Build mapping dictionary
    data = {
        "{{Date of document}}": Date_of_document,
        "{{Duration of Workshop}} ": Duration_of_Workshop,
        "{{Title of Workshop}}": Title_of_Workshop,
        "{{Location of Workshop}}": Location_of_Workshop,
        "{{Date of Workshop}}": Date_of_Workshop,
        "{{Optimistic number}}": Optimistic_number,
        "{{Pessimistic number}}": Pessimistic_number,
        "{{Eng and prof fee}}": Eng_and_prof_fee,
        "{{Total Receipts Optimistic Rs_}}":Total_Receipts_Optimistic,
        "{{Total Receipts Pessimistic Rs_}}":Total_Receipts_Pessimistic,
        "{{Teaching faculty fee}}": Teaching_faculty_fee,
        "{{Eng college student fee}}": Eng_college_student_fee,
        "{{Arai acy student fee}}": Arai_acy_student_fee,
        "{{Total Receipts Optimistic}}": Total_Receipts_Optimistic,
        "{{Total Receipts Pessimistic}}": Total_Receipts_Pessimistic,
        "{{Optimistic Manpower cost Rs_}}": Optimistic_Manpower_cost_Rs_,
        "{{Pessimistic Manpower cost Rs_}}": Pessimistic_Manpower_cost_Rs_,
        "{{Notional_percentage}}": Notional_percentage,
        "{{Optimistic Notional Rs_}}": Optimistic_Notional_Rs_,
        "{{Pessimistic Notional Rs_}}": Pessimistic_Notional_Rs_,
        "{{Optimistic hall charges Rs_}}": Optimistic_hall_charges_Rs_,
        "{{Pessimistic hall charges Rs_}}": Pessimistic_hall_charges_Rs_,
        "{{Honorarium}}": Honorarium,
        "{{Transport Speaker Rs_Approx}}": Transport_Speaker_Rs_Approx,
        "{{Optimistic Honorarium Rs _}}": Optimistic_Honorarium_Rs_,
        "{{Pessimistic Honorarium Rs _}}": Pessimistic_Honorarium_Rs_,
        "{{Optimistic Delegate kits Rs_}}": Optimistic_Delegate_kits_Rs_,
        "{{Pessimistic Delegate kits Rs_}}": Pessimistic_Delegate_kits_Rs_,
        "{{Optimistic Catering Expenses}}": Optimistic_Catering_Expenses,
        "{{Pessimistic Catering Expenses}}": Pessimistic_Catering_Expenses,
        "{{Optimistic Miscellaneous expenses}}": Optimistic_Miscellaneous_expenses,
        "{{ Pessimistic Miscellaneous expenses}}": Pessimistic_Miscellaneous_expenses,
        "{{Optimistic total expense}}": Optimistic_total_expense,
        "{{Pessimistic total expense}}": Pessimistic_total_expense,
        "{{Optimistic Surplus of expense}}": Optimistic_Surplus_of_expense,
        "{{Pessimistic Surplus of expense}}": Pessimistic_Surplus_of_expense,

    }

    replace_placeholders_in_docx(
        template_path="Template.docx",
        output_path="filled_output.docx",
        mapping=data
    )
    print("Done! Saved as filled_output.docx")
```
